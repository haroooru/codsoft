#**DATA VISUALISATION ON TITANIC DATASET**


### The Titanic dataset contains various details about the passengers aboard the ill-fated RMS Titanic, including demographic and travel-related information. Key columns include Survived, which indicates whether a passenger survived (1) or not (0), and Pclass, which represents the passenger's socio-economic status (1st, 2nd, or 3rd class). The Sex column identifies the passenger's gender, and Age provides the age of the passenger, with some missing values. The SibSp and Parch columns show the number of siblings/spouses and parents/children the passenger was traveling with, respectively. Fare details the price paid for the ticket, while Embarked indicates the port where the passenger boarded the ship (Cherbourg, Queenstown, or Southampton). Other columns include Name, which records the full name of the passenger, Ticket, which lists the ticket number, and Cabin, which shows the cabin the passenger was assigned. Overall, the dataset provides rich information that can be analyzed to explore factors affecting survival on the Titanic, such as age, gender, class, and travel companions.

### **HERE FILE PATH IS READ AND PANDAS IS IMPORTED**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "/content/Titanic-Dataset.csv"

titanic_df = pd.read_csv(file_path)

```

#**Here are the labels of the titanic data set:**

```python
print(titanic_df.head())
titanic_df = pd.read_csv(file_path)

print(titanic_df.head())
```

## ***DATA PROCESSING***

### here the missing Age, Embarkation values(from where the ship is borded) are filled using median and mode cabin coloum (where the passengers stayed and room no) is dropped due too many missing values



```python



titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)

titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0], inplace=True)

titanic_df.drop(columns='Cabin', inplace=True)

```

# ***VISUALISATION USING HISTOGRAM***
AGE DISTRIBUTION OF TITANIC PASSENGERS

```python

plt.figure(figsize=(8, 5))
sns.histplot(titanic_df['Age'], bins=30, color='skyblue')
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### AGE BY FREEQUENCY

###this histogram shows the distribution of age of all the titanic passengers on bord this shows that most were between of age between 20 and 40 years,fewer passengers were at the younger and older extremes.This pattern suggests that the majority of travelers were adults, with a smaller number of children and elderly individuals on board.

# **VISUALISATION USING BAR PLOT**

```python

plt.figure(figsize=(6, 4))
sns.countplot(data=titanic_df, x='Survived', palette='viridis')
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

## **SURVIVAL COUNT**

###The bar plot shows the count of passengers who survived as 1 versus those who did not survive as 0 This visualization provides an immediate view of the overall survival rate of titanic passengers.The plot reveals a significant imbalance, with more than 500 passengers not surviving than surviving passengers between 300 and 400 numbers###



#***VISUALISATION USING BARPLOT***

```python

plt.figure(figsize=(8, 5))
sns.countplot(data=titanic_df, x='Pclass', hue='Survived', palette='viridis')
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived')
plt.show()
```

# **SURVIVAL BY PASSENGERS**

### this graph is on survival vs class of passengers where green survived and blue did not From the barplot,it is clear that first-class passengers had the highest survival rate, with a large portion surviving. Second-class passengers had a moderate survival rate, while third-class passengers had the lowest survival rate, with most passengers not surviving. The graph visually emphasizes a direct relationship between class and survival rate, where higher classes correlate with increased chances of survival

# **VISUALISATION USING SCATTER PLOT**

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(data=titanic_df, x='Age', y='Fare', hue='Survived', palette='coolwarm')
plt.title('Fare vs Age (Colored by Survival)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()


```

## **FARE VS AGE**


###This scatter plot is on fare vs age where red dot indicate survived and blue not This scatter plot shows no clear link between age and fare, but younger passengers, especially kids, were more likely to survive. People who paid higher fares had a better chance of survival, with first-class passengers standing out. Older adults (over 60) had a lower survival rate. Overall, age and fare seemed to affect survival, but other factors might have played a role too.

# ***VISUALISATION USING BARPLOT***

```python

plt.figure(figsize=(6, 4))
sns.countplot(data=titanic_df, x='Sex', hue='Survived', palette='Set2')
plt.title('Survival by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')Visualization 5: Survival by Gender (Bar Plot)
plt.legend(title='Survived')
plt.show()
```

# ***VISUALISATION USING BARPLOT OF MEAN SURVIVAL***

```python

plt.figure(figsize=(8, 5))
sns.barplot(data=titanic_df, x='Pclass', y='Survived', ci=None, palette='viridis')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

##**SURVIVAL RATE VS PASSENGERCLASS**

###The bar plot shows the survival rate for each passenger class on the Titanic. It displays the average survival rate for passengers in first, second, and third class. The plot uses the 'viridis' color palette to differentiate the classes, and it shows that first-class passengers had the highest survival rate, followed by second-class passengers, with third-class passengers having the lowest survival rate. This suggests that passengers in higher classes were more likely to survive the disaster.








# ***VISUALISATION USING BOXPLOT***

```python

plt.figure(figsize=(8, 5))
sns.boxplot(data=titanic_df, x='Survived', y='Age', palette='Set2')
plt.title('Age Distribution by Survival')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()
```

##**AGE VS SURVIVAL**

###The box plot shows the age distribution for passengers who survived and those who didnt. It compares the ages of survivors (1) and non-survivors (0). The plot reveals that most survivors were younger, with the median age for survivors being lower than for non-survivors. There are also more younger passengers in the survival group, while the non-survival group has a wider age range, including more older passengers. This suggests that younger people had a higher chance of survival on the Titanic.
